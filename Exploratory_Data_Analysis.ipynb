```python
# Read in the document-term matrix
import pandas as pd

data = pd.read_pickle('dtm_tsai.pkl')
data = data.transpose()
data.head()
```

```python
data['President_Tsai'].sort_values(ascending = False).head(15)
```

```python
# Find the top 50 words said by the president
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(40)
    top_dict[c]= list(zip(top.index, top.values))

top_dict
```
